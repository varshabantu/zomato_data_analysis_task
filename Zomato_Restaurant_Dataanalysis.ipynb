# 2. Data Cleaning


```python
#Remove duplicates records
df = df.drop_duplicates()
```

```python
df = df.drop(["Unnamed: 0","rest_type","menu_item","location","phone","url","reviews_list","address"],axis=1)
df
```

```python
# Rename listed_in_type to rest_type
df.rename(columns={"listed_in(type)":"rest_type"}, inplace=True)
#Rename listed_in_city to location
df.rename(columns={"listed_in(city)":"location"}, inplace=True)
```

```python
#Clean rate column

df["rate"] = df["rate"].replace(["NEW", "-"], np.nan)
# removing /5 data
df["rate"] = df["rate"].astype(str).str.split("/5").str[0]
# For rate column converting into numeric values
df["rate"] = pd.to_numeric(df["rate"], errors="coerce")
df["rate"].unique()
df["rate"] = df["rate"].fillna(df["rate"].median())
df["rate"].unique()
```

```python
#Clean Cost column

# Replacing , to empty string in cost column/ removing commas from the data
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].replace(",","")
#cost columns to numeric values
df["approx_cost(for two people)"] = pd.to_numeric(df["approx_cost(for two people)"], errors="coerce")
df["approx_cost(for two people)"].unique()
```

```python
df.isnull().sum()
```

```python
# Handling dish_liked column have many missing values
df["dish_liked"] = df["dish_liked"].fillna("Not Available")
df["dish_liked"].unique()

# cuisines have very few missing values (string)
# df["cuisines"] = df["cuisines"].fillna(df["cuisines"].mode()[0])
```

```python
#Handling cuisines column have very few missing values
df["cuisines"] = df["cuisines"].fillna(df["cuisines"].mode()[0])
df["cuisines"].unique()
```

```python
#Handling Rate columns
df["rate"].fillna(df["rate"].median(), inplace=True)
df["rate"].unique()
```

```python
#Handling missing values for votes column
df["votes"].fillna(df["votes"].median(), inplace=True)
df["votes"].unique()
```

```python
#Handling missing values for approx_cost column
df["approx_cost(for two people)"].fillna(df["approx_cost(for two people)"].median(), inplace=True)
df["approx_cost(for two people)"].unique()
```

```python
df.info()
```

```python
df["online_order"] = df["online_order"].replace(["Yes", "No"])
df["online_order"].unique()
```

```python
df["votes"].unique()
```

# Task 2: Univariate Analysis

**Perform the following analysis with proper visualizations:**

1. Top 5 locations with highest number of restaurants


```python
# Count how many times each location appears
top_locations = df["location"].value_counts().head(5)

# Create a new figure with given size
plt.figure(figsize=(6,4))

# Plot the top 5 locations as a bar chart
plt.bar(top_locations.index, top_locations.values,color="Red")

# Set x-axis label
plt.xlabel("Location")

# Set y-axis label
plt.ylabel("Number of Restaurants")

# Set title for the graph
plt.title("Top 5 Locations with Highest Number of Restaurants")
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

The top locations contain a very high number of restaurants, showing strong market competition.

2. Distribution of restaurant ratings.

```python
# Create a hist plot
plt.figure(figsize=(6,4))
plt.hist(df["rate"], bins=20, color="Red")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

Most restaurants fall in the medium rating range (around 3.5 to 4.0).
Very few restaurants have extremely low or extremely high ratings.

3. Percentage of restaurants offering online orders.

```python
# Calculate percentage values in online_order column
online = df["online_order"].value_counts(normalize=True) * 100
plt.figure(figsize=(5,5))
plt.pie(online,labels=online.index,autopct="%1.1f%%",shadow=True,explode=(0,0))
plt.title("Percentage of Restaurants Offering Online Order")
plt.show()
```

A large proportion of restaurants offer online ordering facilities.
This indicates that online ordering has become an important service feature.

4. Distribution of average cost for two.

```python
# Calculate the average cost.
sns.histplot(data=df,x="approx_cost(for two people)",kde=True)
plt.xlabel("Approx Cost (for two people)")
plt.ylabel("Count")
plt.title("Distribution of Approx Cost for Two People")
plt.show()


```

Most restaurants are concentrated in the low to medium price range.
Only a small number of restaurants belong to the high-cost category.

# Task 3 : Bivariate Analysis

 Analyze the relationship between:

1. Rating vs Cost

```python
#Rating vs Cost

# Create a scatter plot for average cost and rating
plt.figure(figsize=(6,4))
plt.scatter(df["approx_cost(for two people)"], df["rate"], color="Red")
plt.xlabel("Average Cost for Two")
plt.ylabel("Rating")
plt.title("Rating vs Cost")
plt.show()
```

2. Rating vs Online Order

```python
# Rating vs online Order

plt.figure(figsize=(6,4))
sns.barplot(x="online_order",y="rate",data=df)
plt.xlabel("Online Order")
plt.ylabel("rate")
plt.title("Rating vs Online order")
# plt.xticks(rotation=90)
plt.show()
```

3. Votes vs Rating


```python
# Votes vs Rating

# Create a correlation matrix for numeric columns

corr_matrix = df[['votes', 'rate']].corr()

plt.figure(figsize=(5,4))
sns.heatmap(corr_matrix, annot=True, cmap="rainbow")
plt.title("Correlation Heatmap: Votes vs Rating")
plt.show()
```

Yes, We can see the higher cost leads to higher rate
